¿Qué es la tasa de retorno en la educación? El cambio en el ingreso por estudiar un año más. 
Dos hechos: 
    * los retornos a la educación son más altos en países con menor desarrollo. Movilidad social: cada generación tenga mayor calidad de vida, mayor ingreso, mayor etc. Mayor movilidad social, entonces distribución más egalitaria. 
Los resultados que otros autros autores encontraron no verifica lo anterior.
    
    * Muy probablemente los estudios empíricos en estudios a la educación estaban sobreestimando el efecto. Hipótesis: los autores han sobreestimado los retornos a la educación cuando han hecho análisis empírico en esos países. 
    * Hipótesis: omitieron antecedentes de los padres y características no observables. El problema es que hay al menos una característica que está en el error y está correlacionada con ambas la variable dependiente e independiente de interés.
    * El capital humano contiene a los años de la escolaridad más la calidad de la educación, más el entorno familiar, tu inteligencia y habilidades. Un individuo sin habilidades difícilmente llegará a años altos de escolaridad. 

$$\ln(w) = \alpha + X\beta + \theta S + \epsilon$$

$$\mathbb{E}\Bigg[\frac{\partial \ln(w)}{\partial S}\Bigg] = \mathbb{E}\Bigg[ \hat{\theta} + \frac{\partial \epsilon}{\partial S}\Bigg] = \theta + sesgo$$

Es decir, la habilidad puede estar en el error. También nos importa la capacidad cognitiva adquirida en vida. 

¿Cuál podría ser la proxy de la habilidad? Escolaridad de los padres.

###Formalicemos el modelo

$$y_i = \beta_0 + \beta_s S_i + \epsilon_i,$$
donde $\epsilon_i$ representa a la habilidad no obserbada; es decir, $\epsilon_i \equiv Ability_i \equiv \beta_a A_i + u_i$, donde $u_i$ respresenta características no observables, un ruido blanco.

$$ \frac{\partial y}{\partial S} = \beta_s + \beta_a\frac{\partial A}{\partial S} = \beta_s + \beta_a\sigma_s \Longrightarrow \hat{\beta_s} = \beta_s + \beta_a\sigma_s = \beta_s$$

[1] $$\ln(y) = X\beta + \gamma S + \epsilon$$
[2] $$\ln(y) = X\beta + \gamma S_i + \gamma_{father}^{i} S_F + \gamma_{mother}S_m + \epsilon$$
[3] $$\ln(y) = X\beta + \gamma S + \theta_{wf} S_{wf} + \theta_{wm} S_{wm} + \epsilon$$

Sólo educación del cóngyuge:

[4] $$\ln(y) = X\beta + \gamma S + \theta_{w} S_{w} + \epsilon$$

Todas:

[4] $$\ln(y) = X\beta + \gamma S + [\gamma S_{f} + \gamma S_{m}] + \theta_w S_w + [\theta_{wf} S_{wf} + \theta_{wm} S_{wm}] + \epsilon$$

(1) $$\frac{\partial y}{\partial s} = \beta_s + \frac{\partial \beta_a A_i}{\partial S} \Longrightarrow \frac{\partial y}{\partial s} = \beta_s + \sigma_{A_{is}}, $$

donde $A_i = \gamma_{f} S_{f} + A_{i}^m$

(2) $$\frac{\partial y}{\partial s} = \beta_s + \frac{\partial A_i^{\mu}}{\partial S} \Longrightarrow \frac{\partial y}{\partial s} = \beta_s + \sigma_{A_{is}^{\mu}}$$

Así, $\sigma_{A_{is}} > \sigma_{A_i^{\mu}} \Longrightarrow \hat{\beta_s^1} > \hat{\beta_s^2}$
