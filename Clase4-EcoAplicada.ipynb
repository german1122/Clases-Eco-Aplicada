#Estimación Empírica de funciones de oferta laboral

###[1] Decisión de participación laboral (trabajo o no trabajo): análisis de margen extensivo;

###[2] Decisión sobre las horas trabajadas y diferentes salarios: análisis de margen intensivo

[1] Participación de la fuerza laboral ignora la información de las horas trabajadas. $h_i$ se colapsa en 2 valores

$$ PFL =
  \begin{cases}
    1       & \quad \text{Sí trabajan} \\
    0  & \quad \text{no trabajan}\\
  \end{cases}
$$


[2]$$h_i = \mathbf{X\beta} + \mathbf{u},$$
donde $h_i$ nos dice si trabaja o no; $\mathbf{X}$ es una función de características observables y $\mathbf{u}$ representa características no observables del individuo.

En una distribución normal, $h_i$ colapsa en dos valores: 0 y 1, por lo que lo denominamos modelo de probabilidad lineal. Donde,

$$\mathbb{E}[h_i | \mathbf{X}] = \mathbf{X\boldsymbol\beta} + \mathbb{E}[\mathbf{u} | X] = 1·P[h_i = 1 | \mathbf{X}] + 0·P[h_i = 0 | \mathbf{x}] = \mathbf{X\beta}$$

Por tanto, si corremos [2] por MCO y estimamos $\hat{h_i} = \mathbf{X\boldsymbol\beta}$ nos da la probabilidad estimada de participación en la fuerza laboral, dado $\mathbf{X}$.

(Primer problema) Nada nos garantiza que los valores de $\mathbf{X}\boldsymbol\beta$ estén entre cero y uno. Ese es un problema, una desventaja. Es decir, puedo estimar probabilidades arriba de uno y abajo de cero.

(Segundo problema) Tenemos heteroscedasticidad.
$$h_i = \mathbf{X\boldsymbol\beta} + u_i$$

Posibles valores de $u_i$ y $f(u_i)$:
$$\begin{matrix}
  & u_i & f(u_i)  \\
 u_i = 1 & 1-\mathbf{X\boldsymbol\beta} & P[h_i = 1|\mathbf{X}] = \mathbf{X\boldsymbol\beta} \\
 u_i = 0 & -\mathbf{X\boldsymbol\beta} & 1-P[h_i = 1|\mathbf{X}]=1-\mathbf{X\boldsymbol\beta}
 \end{matrix}$$

$$\Longrightarrow \mathbb{Var}(u_i) = \mathbb{E}[u_i^2] - \mathbb{E}^2[u_i] = \mathbb{E}[u_i^2]$$

$$\mathbb{E}[u_i^2] =\sum_{i=0}^1 u_i^2f(u_i)  = x_i\boldsymbol\beta(1-x_i\boldsymbol\beta)^2 + (1-\boldsymbol\beta x_i)·(-x_i\boldsymbol\beta)^2 = (1-x_i\beta)[x_i\beta-(x_i\beta)^2+(x_i\beta)^2] =x_i\boldsymbol\beta(1-x_i\boldsymbol\beta)= \hat{h_i}(1-\hat{h_i}),$$
lo cual está en función de $\mathbf{X}$.

Tenemos que encontrar una solución donde los valores de $h_i^*$ estén acotados entre cero y uno cuando $\mathbf{X\boldsymbol\beta} \in (-\infty, \infty)$. Solución: variable latente (concepto).

[4] $$h_i^* = \mathbf{X\boldsymbol\beta} + \mathbf{u}$$

[5] $$ PFL =
  \begin{cases}
    1       & \text{si} \quad h_i^* \geq 0 \\
    0   &  \text{si} \quad h_i^* < 0\\
  \end{cases}$$
A $h_i^*$ lo pensamos como las horas deseadas de trabajo.

De [4] y [5], 
$$\mathbb{Prob}[h_i = 1 | \mathbf{X}] = \mathbb{Prob}[h_i^* > 0 | \mathbf{X}] = \mathbb{Prob}[\mathbf{X\boldsymbol\beta} + \mathbf{u} > 0] = \mathbb{Prob}[u > -\mathbf{X\boldsymbol\beta}].$$
Dado lo anterior, si la distribución es simétrica,

$$\mathbb{Prob}[h_i = 1 | \mathbf{X}] = \mathbb{Prob}[\mathbf{u} < \mathbf{X\boldsymbol\beta}].$$

[6] $$\mathbb{Prob}[h_i = 1 | \mathbf{X}] = \int_{-\infty}^{\mathbf{X\boldsymbol\beta}} f(u_i)df \equiv F[\mathbf{X\boldsymbol\beta}] $$
$$\mathbb{Prob}[h_i=0 | \mathbf{X}] = 1-F[\mathbf{X\boldsymbol\beta}]$$

Los estimadores de $\beta$ nos permitirán estimar la probabilidad acotada entre cero y uno del individuo de participar en la fuerza laboral. Método: máxima verosimilitud ($u_i \sim iid$), donde $F$ es una función de distribución (ya sabe cómo es su gráfica teórica, con rango entre cero y uno).

$$Máx \quad \mathbb{P}[Y_1 = y_1, Y_2 = y_2, \cdots Y_n = y_n] = Máx \quad \prod_{y_i = 0} [1-F(\mathbf{X\boldsymbol\beta})]\prod_{y_i = 1} [F(\mathbf{X\beta})] $$

Así, sea

$$L = \prod_{y_i = 0}^{n} [1-F(\mathbf{X\boldsymbol\beta})]^{1-y_i} [F(\mathbf{X\boldsymbol\beta})]^{y_i}$$

Tomando logaritmos, el problema se convierte en:

$$Máx \quad \ln[L] = \sum_{i}^n (1-y_i)\ln[1-F(\mathbf{X\boldsymbol\beta})] + y_i \ln[F(\mathbf{X\boldsymbol\beta})]$$

F.O.C.

$$\frac{\partial ln(L)}{\partial \beta} = \sum_{i}^n \Bigg [\frac{y_i f_i}{F_i} + (1-y_i)\frac{-f_i}{1-F_i}\Bigg]_{x_i = 0}, $$

donde $F_i \equiv F(x_i\boldsymbol\beta)$ & $f_i \equiv f(x_i\boldsymbol\beta)$, $\hat{\boldsymbol\beta} = \boldsymbol\beta_{MLE}$.

##Modelo Probit

Si suponemos que $u_i \sim N[0, \sigma_u^2] \Rightarrow $ Modelo Probit. Así, $\frac{u_i}{\sigma_u^2} \sim N[0,1]$ es un modelo probit de una distribución normalizada.

$$Máx \quad \ln[L] = \sum_{i}^n (1-y_i)\ln\Bigg[1-F\Bigg(\frac{\mathbf{X\boldsymbol\beta}}{\sigma}\Bigg)\Bigg] + y_i \ln\Bigg[F\Bigg(\frac{\mathbf{X\boldsymbol\beta}}{\sigma}\Bigg)\Bigg],$$

donde $\boldsymbol\beta_{MLE}^* = \frac{\boldsymbol\beta}{\sigma_u},$ donde asumimos que $\sigma^2 = 1$. Recuerden que la integral de una normal estándar no está definida, por lo que habría que utilizar métodos numéricos de aproximación.

##Modelo Logit

$u_i \sim$ distribución logística.


$$F(\mathbf{X\boldsymbol\beta}) = \frac{\exp(X\boldsymbol\beta)}{1+\exp(X\boldsymbol\beta)} =
1-F(\mathbf{X\boldsymbol\beta}) = 1-\frac{\exp(X\boldsymbol\beta)}{1+\exp(X\boldsymbol\beta)} = \frac{1}{1+\exp(\mathbf{X\boldsymbol\beta})}$$
 

En la práctica, los estimadores probit y logit son muy parecidos, salvo en las colas.

##Modelo de Probabilidad lineal
$$\mathbb{E}[Y | X] = \mathbb{P}[Y = 1 | X] = X\beta $$
$\beta_{MCO}$ sí es consistente: $\mathbb{E}[\mathbf{u'X}] = 0$.
$$\frac{\partial (y = 1 | \mathbf{X})}{\partial \mathbf{X}} = \frac{\partial \mathbf{X\beta}}{\partial \mathbf{X}} = \beta$$

##¿Logit o Probit?

$$\mathbb{E}[\mathbf{Y} | \mathbf{X}] = \mathbb{P}[\mathbf{Y} = 1| \mathbf{X}] = F(\mathbf{X\beta})$$

¿Cuál es la dificultad de usar un probit?:
    $$\frac{\partial \mathbb{P}[\mathbf{Y} = 1| \mathbf{X}]}{\mathbf{X}} = \frac{\partial F(\mathbf{X\beta})}{\partial \mathbf{X\beta}} ·\frac{\partial \mathbf{X\beta}}{\partial\mathbf{X}}  = f(\mathbf{X\beta})\beta$$

##Oferta Laboral

Margen intensivo: ¿Cómo afectan a las x's cuántas horas trabajo?
$$h_i = \mathbf{X\beta} + \mathbf{u}$$
$h_i \equiv $ cantidad de horas trabajadas.

Empíricamente, sólo observamos $h_i \geq 0$. Es decir, no observamos las decisiones de los que no trabajan y tampoco sus "salarios", dado el costo de oportunidad que enfrentan respecto del ocio.

Problema de selección de muestra: sólo vemos las decisiones de aquellos que entraron a la fuerza laboral. Dentro de la distribución, lo observado no es extrapolable al resto de la población. Si no adevertimos este problema de MCO, 

$$h_i = \mathbf{X\beta} + \mathbf{u} \Longrightarrow \mathbb{E}[h_i | \mathbf{X}] = \mathbf{X\beta} + \mathbb{E}[\mathbf{u}] = \mathbf{X\beta} $$
Lo anterior sucede en toda la población. Sin embargo, si tenemos un problema de selección de muestra, no necesariamente $\mathbb{E}[\mathbf{u} | \mathbf{X}, \quad h_i > 0 ] \neq 0$

$$\mathbb{E}[h_i | \mathbf{X} \quad h_i > 0 ] = \mathbf{X\beta} + \mathbb{E}[\mathbf{u} | \mathbf{X}, \quad h_i > 0 ]$$

$$\frac{\partial \mathbb{E}[h_i | \mathbf{X} \quad h_i > 0 ]}{\partial\mathbf{X}} = \frac{\partial\mathbf{X\beta}}{\partial \mathbf{X}} + \frac{\mathbb{E}[\mathbf{u} | \mathbf{X}, \quad h_i > 0 ]}{\partial \mathbf{X}} = \beta + sesgo$$

Por tanto, el estimador por MCO es inconsistente e insesgado.

##Modelo Tobit [Tobin (1958)]
Soluciona el problema de muestra censurada. Tengo $x_i$ para todos incluyendo para las $h_i>0$ y $h_i = 0$.
###Muestra truncada
Sólo observamos $\mathbf{X}$ cuando observamos $h_i \geq 0$.

Objetivo: incluir en el modelo la información de los que trabajan y los que no trabajan.
La información de los trabajadores:

$$\mathbb{P}[h_i = 0 |\mathbf{X}] = 1- F\Bigg(\frac{\mathbf{X\beta}}{\sigma}\Bigg),$$

donde lo anterior representa la probabilidad de mantenerse fuera del mercado laboral.

La contribución de los trabajadores:
    [13]$$f(h_i | h_i > 0)·\mathbb{P}(h_i>0)$$

###Teorema de Bayes

  [14]
    $$f(h_i | h_i > 0) = \frac{f(h_i, h_i > 0)}{f(h_i > 0)} \equiv \frac{f(h_i, h_i > 0)}{\mathbb{P}(h_i>0)}$$

Sustituyendo [13] y [14],


Su contribución al modelo es simplemente sus horas trabajadas.

$$f(h_i | h_i > 0) = \frac{f(h_i, h_i > 0)}{\mathbb{P}(h_i>0)}·\mathbb{P}(h_i>0) = f(h_i | h_i > 0)$$

El propósito del modelo Tobit es maximizar la siguiente función de máxima verosimilitud:
    
$$L = \prod_{h_i > 0}\Bigg[F\Bigg(\frac{h_i-\mathbf{X\beta}}{\sigma}\Bigg)\Bigg]\prod_{h_i=0}\Bigg[1-F\Bigg(\frac{\mathbf{X\beta}}{\sigma}\Bigg)\Bigg] $$

Ventaja: toma en cuenta el proceso de selección de muestra mediante una modelación de este proceso;

Desventajas: usa las mismas x's para la decisión de participar y de horas trabajadas;
sólo sirve para resolver el problema de muestra censurada;
los x's tienen el mismo efecto para estimar el margen extensivo e intensivo.
