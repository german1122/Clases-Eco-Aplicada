#Concepto KH: Capital Humano
##Modelo de Mincer (1958)
###Concepto: Valor Presente (Perpetuidad)
    - Tiempo discreto y continuo;

##Modelo de Mincer

$$H = CI + CD, $$
donde 

$CI \equiv$ costos indirectos en educación (en Valor Presente);

$CD \equiv$ costos directos en educación (en Valor Presente);

Retorno del capital humano:
$VP[R(H)]$, donde $R$ es la función de ingresos laborales en función de la inversión en $H \Longleftrightarrow (KH) $.

$H \equiv$ costo del capital humano expresado en dinero;

Si $H < VP[R(H)]$, el individuo escogerá un nivel de $H^*$ hasta que 

[3]$$H = VP[R(H)]$$

Estamos suponiendo que toda inversión en capital humano se da porque el individuo espera un retorno. Pero eso no es cierto. Podría ser que el individuo decida invertir en capital humano por mero gusto. También hay partes de la inversión en capital humano que puede ser consumo. Lo anterior me podría dar un retorno en mi función de utilidad indirecta, pero esto lo simplificaremos por ahora. ¿Cómo podemos calcular esto?

[4]$$VP[R(H)] =  \sum_{t=1}^T \frac{R_t(H)}{(1+i_t)^t}$$ ,
donde $T$ es el momento en el que te jubilas. 

$T \equiv$ horizonte productivo;
$i \equiv$ tasa de interés (real);
$R(H)_t \equiv$ ingreso laboral en $t$

Si suponemos que la tasa de interés ($i$) es fija,
[4]$$VP[R(H)] =  \sum_{t=1}^T \frac{R_t(H)}{(1+i)^t}$$
Si multiplicamos con $(1+i)$, 

$$(1+i)VP[R(H)] = \sum_{t=0}^{T-1} \frac{R_t(H)}{(1+i)^t}$$

Ahora substraemos la primera ecuación de la segunda, 

[5]$$VP[R(H)] -(1+i)VP[R(H)] = R(H)\Big[1-\frac{1}{(1+i)^T}\Big] \Longrightarrow VP[R(H)] = \frac{R(H)}{i}\Big[1-\frac{1}{(1+i)^T}\Big]$$


Ejemplo: una persona tiene un sueldo promedio de \$ 10,000 y piensa trabajar unos 30 años, donde la tasa de interés es de 50% anual. Así,

$$VP[R(Primaria)] =\frac{10,000}{0.5}\Big[1-\frac{1}{(1.5)^{30}}\Big] = 19.9$$

Con una perpetuidad, $T \rightarrow \infty$, entonces
$$VP[R(H)]_{T \rightarrow \infty } = \frac{R(H)}{i} = 20,000$$

  $$VP(s, r) = VP(s+d, r), $$
  
  $r^* \equiv \rho(s, s+d)$, donde $r$ es la tasa a ajustar.
  

Entonces, si definimos el $VP$ del ingreso laboral de vida útil:

$$VP(s, r) = \int_0^T \psi(s, x)e^{-r(s+d)}dx$$, 

donde $\psi \equiv$ función del ingreso laboral que depende de la escolaridad y la experiencia.
$x \equiv t-s-6$.
La tasa de retorno $\rho(s, s+d)$ de $d$ años de escolaridad adicional. Así, 

$$\frac{\partial V[\psi (s, x)]\%}{\partial s} = \rho(s,s+d)$$

[9]$$\hat{\rho}(s,s+d) = \frac{1}{d}\Bigg\{\ln\Bigg[\int_0^T \psi (s+d,x) e^{-r(s+d, x)}dx\Bigg]-\ln\Bigg[\int_0^T \psi (s,x) e^{-r (s, x)}dx\Bigg]\Bigg\}$$

Lo anterior se puede calcular mediante métodos numéricos.

Para poder obtener una expresión analítica vamos a asumir una función de ingresos débilmente separable:
    
[10]    $$y = \psi (s,x) = f(s)·g(x) \Longrightarrow $$
[11]    $$VP(s,r) = f(s)e^{-rs} \int_0^T g(x) e^{-rx}dx$$

Así, si juntamos [9], [10] y [11], obtenemos

$$\hat{\rho}(s,s+d) = \frac{1}{d}\Bigg\{\ln\Bigg[f(s+d) e^{-r(s+d)}\int_0^T g(x)e^{-rx}dx\Bigg]-\ln\Bigg[f(s) e^{-r(s)}\int_0^T g(x)e^{-rx}dx\Bigg]\Bigg\} = \frac{1}{d}\Bigg[ \ln f(s+d)e^{-r(s+d)} - \ln f(s) e^{-rs}\Bigg]$$

$$\hat{\rho}(s,s+d) = \frac{1}{d}\Bigg[ \ln f(s+d) - \ln f(s) -rd\Bigg]$$
Si $rd=0 \Longrightarrow$ 

$$\hat{\rho}(s+d) = \frac{1}{d}\Bigg[ \ln f(s+d) - \ln f(s)\Bigg]$$
