#Exogeneidad



Si 

[Modelo Poblacional]$$Y = X\beta + \epsilon$$

$$\mathbb{E}[X'\epsilon] = 0$$

Estimador óptimo: aquél que replique a la realidad. Sea 

[Modelo muestral]

$$y = xb + e \Longrightarrow \mathbb{E}[X'\epsilon] = 0 \Longrightarrow x'e = 0 = x'[y-xb] = 0 \Longrightarrow x'y -x'xb = 0 \Longrightarrow \hat{b} = (x'x)^{-1}x'y = \widehat{b_{MCO}}$$

$\widehat{b_{MCO}} \Rightarrow$ consistente e insesgado.

$$\hat{\beta} = (X'X)^{-1}(X'Y) = (X'X)^{-1}X'(X\beta+\epsilon) = (X'X)^{-1}X'X\beta +(X'X)^{-1}X'\epsilon = \beta + (X'X)^{-1}X'\epsilon \Longrightarrow \mathbb{E}[\hat{\beta}] = \beta + (X'X)^{-1}\mathbb{E}[X'\epsilon] = \beta$$

En presencia de endogeneidad, $\mathbb{E}[X'\epsilon] \neq 0 \Longrightarrow \mathbb{E}[\beta] = \beta + sesgo$

Sea $y^* = x^*\beta + \epsilon$ el modelo verdadero, donde

$$y = y^* + v,$$

$$x = x^* + \mu,$$

donde $y$ y $x$ son cada una medidas con error. $v$ nos genera varianza, pero no causa sesgo, mientras que $\mu$ sí causa sesgo.
Así, 

$$y - v = (x-\mu)\beta + \epsilon \Longrightarrow y = x\beta + \epsilon + v - \mu \beta = x\beta + \tilde{\epsilon}$$



Por tanto, 

$$\mathbb{E}[x'\tilde{\epsilon}] = \mathbb{E}[x'\epsilon] + \mathbb{E}[x'v] -\mathbb{E}[x'\mu]\beta = -\mathbb{E}[x'\mu]\beta \Longrightarrow \mathbb{E}[\widehat{\beta_{MCO}}] = \beta + \mathbb{E}[(x'x)^{-1}x'\epsilon^*]$$

Prueba en el caso $K = 1$:

$$b_{MCO} = \frac{\widehat{Cov(y,x)}}{\widehat{Var(x)}} = \frac{\widehat{Cov(y^*+v,x^*+\mu)}}{\widehat{Var(x^*+\mu)}} = \frac{\widehat{Cov(y^*,x^*)} +\widehat{Cov(y^*,\mu)} + \widehat{Cov(v,x^*)} + \widehat{Cov(\mu,v)}}{\widehat{Var(x^*)} + \widehat{Var(\mu)} + 2\widehat{Cov(v,\mu)}}$$

Así, 

$$b_{MCO} = \frac{\widehat{Cov(y^*,x^*)}}{\widehat{Var(x^*)} + \widehat{Var(\mu)} }· \frac{\widehat{Var(x^*)}}{\widehat{Var(x^*)}} = \frac{\widehat{Cov(y^*, x^*)}}{\widehat{Var(x^*)}}·\frac{\widehat{Var(x^*)}}{\widehat{Var(x^*)} + \widehat{Var(\mu)} } $$ 


Por tanto, 

$$plim[b_{MCO}] = \frac{Cov(y^*, x^*)}{Var(x^*)}·\frac{Var(x^*)}{Var(x^*) + Var(\mu)}, $$ 


donde el denominador del segundo término de la multiplicación es el sesgo por subestimación.

##Variables Instrumentales

Posible solución: $\mathbb{E}[x'\epsilon] \neq 0$, por lo que debemos encontrar una $z$ tal que $\mathbb{E}[z'\epsilon] = 0$.
Así, 

[modelo muestral] $$y = xb + e \Longrightarrow z'e = 0 = z'[y-xb] \Longrightarrow \widehat{b_{IV}} = (z'X)^{-1}z'y, $$

donde 

$$\widehat{b_{IV}} = (z'x)^{-1}z'(xb+e) = b + (z'x)^{-1}z'e  \Longrightarrow \mathbb{E}[\widehat{b_{IV}}] = \beta + (Z'X)^{-1}\mathbb{E}[Z'\epsilon] = \beta$$


La condición de IV que garantiza la consistencia:
    
$$\mathbb{E}[z'e] = 0$$
    
Segunda condición: 

$$z'x = 0$$

Así, 

$$\widehat{b_{IV}} = (z'x)^{-1}z'y = \beta + (z'x)^{-1}z'e \Longrightarrow Var(\widehat{b_{IV}}) = Var(\beta) + Var[(z'x)^{-1}z'e] = (Z'X)^{-1}Z\sigma^2 Z(X'Z)^{-1} = \sigma^2 (Z'X)^{-1}Z'Z(X'Z)^{-1}$$
