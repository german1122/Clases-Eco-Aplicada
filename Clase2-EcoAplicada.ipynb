### *Condiciones de agregación

[RP]$$Px-(T-l)w = y$$

Supuestos: En no saciedad siempre estamos consumiendo sobre restricción presupuestal.

Diferenciando totalmente respecto de  $y$:

$$Pdx + xdP-Tdw + wdl - ldw = dy \Longrightarrow P\frac{\partial x}{\partial y} + w\frac{\partial l}{\partial y} = \frac{\partial y}{\partial y} = 1$$

$$P\frac{\partial x}{\partial y} + w\frac{\partial l}{\partial y} = 1$$ (Condición de agregación de Engel)

$$P\frac{\partial x}{\partial y}\frac{ y}{ y}\frac{ x}{ x} + w\frac{\partial l}{\partial y}\frac{ l}{ l}\frac{ y}{ y} = 1$$

$$\frac{Px}{y}\frac{ \partial x}{\partial y}\frac{ y}{ x} + \frac{ wl}{ y}\frac{ \partial w}{ \partial y}\frac{ y}{ w} = 1$$

[23'] $$\frac{ Px}{y} \epsilon_{xy} + \frac{wl}{y}\epsilon_{ly} = 1$$

Diferenciando totalmente: P, w (Condición de Cournot)
    $$P\frac{\partial x}{\partial y} + w\frac{\partial l}{\partial P}+x = 0$$
    $$P\frac{\partial x}{\partial w} + w\frac{\partial l}{\partial w}-[T-l] = 0$$

###Homogeneidad

Las demandas $x^*$ y $l^*$ son homogéneas de grado cero. 

Definición matemática: En general, se dice $g(x)$ es homogénea de grado $k$ si: [25]
$$ g(tx) = t^k g(x)$$

Dif[25]: t [26]
    $$\sum g_i x_i = k t^{k-1} g(x),$$
    donde $g_i = \frac{\mathrm{d} g}{\mathrm{d} x_i}$

Si g(·) es homogénea de grado cero,
$$\sum g_i z_i =0$$

Aplicando restricción de Marshall,
$$x(tP, tw, ty) = t^k x(p,w,y)$$
sabemos que $x^*$ es de grado cero por CPO

[27]$$P\frac{\partial x}{\partial P} + w\frac{\partial x}{\partial w}+y\frac{\partial x}{\partial y} = 0$$

Sust Slutsky:
   $$ P\frac{\partial x}{\partial P}\Bigg|_{dU = 0} +w\frac{\partial x}{\partial w}\Bigg|_{dU = 0}+\frac{\partial x}{\partial y}[y-px+w(T-l)] = 0$$
   
   

Como $y-px+w(T-l) = 0$, 
 
 $$ \Longrightarrow P\frac{\partial x}{\partial P}\Bigg|_{dU = 0} +w\frac{\partial x}{\partial w}\Bigg|_{dU = 0}=0$$

[28(Por supuesto)]:
    $$w\frac{\partial l}{\partial w}\Bigg|_{dU = 0} +P\frac{\partial l}{\partial P}\Bigg|_{dU = 0}=0$$

##II. Enfoque de utilidad indirecta
[9] $l^* = l(P, w, y,A, \epsilon)$

[10] $x^* = x(P, w, y,A, \epsilon)$

$$U(x^*, l^*) = U[x^*(P, w, y,A, \epsilon), l^*(P, w, y,A, \epsilon)] = V[P, w, y, A, \epsilon]$$

es el valor monetario de mi felicidad cuando [30] consumo mi canasta óptima. 

##Efecto de un $\Delta$ en el ingreso:
$$\frac{\partial V}{\partial y} = \frac{\partial U}{\partial x^*}\frac{\partial x^*}{\partial y} + \frac{\partial U}{\partial l^*}\frac{\partial l^*}{\partial y}$$


Sabemos que $U_x = \lambda P$ & $U_l = \lambda w$ 

$$\Longrightarrow \frac{\partial V}{\partial y} = \lambda\Bigg[P\frac{\partial x^*}{\partial y} + w\frac{\partial l^*}{\partial y}\Bigg], $$

por condición de Engel

[31]$$ \Longrightarrow \frac{\partial V}{\partial y} = \lambda > 0$$
$$ \lambda = \frac{U_x}{P}>0,$$ por la condición de no saciedad.

##Efecto de un $\Delta$ en precio:
$$\frac{\partial V}{\partial P} = U^*_x\frac{\partial x^*}{\partial P} + U^*_l\frac{\partial l^*}{\partial P}, $$ por condiciones de Cournot.

$$\frac{\partial V}{\partial P}  = \lambda\Bigg[\frac{\partial x^*}{\partial P}P + \frac{\partial l^*}{\partial P}w\Bigg], $$ por CPO

[32]$$\frac{\partial V}{\partial P}  = -\lambda_{+} x_{+} \leq 0$$ 

Por otro lado,
$$\frac{\partial V}{\partial w} = U^*_x\frac{\partial x}{\partial W} + U^*_l\frac{\partial l^*}{\partial W}, (CPO) $$

$$\frac{\partial V}{\partial P}  = \lambda\Bigg[P\frac{\partial x}{\partial w} + w\frac{\partial l}{\partial w}\Bigg]$$

Diferenciando totalmente: P, w, condición de Cournot
    $$P\frac{\partial x}{\partial P} + w\frac{\partial l}{\partial P} + x = 0$$
    $$P\frac{\partial x}{\partial w} + w\frac{\partial l}{\partial w} - (T-l) = 0$$

$$\frac{\partial V}{\partial w} = \lambda (T-l)$$
$$ \frac{\partial V}{\partial w} = \lambda h^* \geq 0, [33]$$

##Identidad de Roy

$$\frac{\frac{\partial V}{\partial P}}{\frac{\partial V}{\partial y}} = \frac{-\lambda x}{\lambda} = -x^*(P,w,y, \epsilon, A)$$

obtengo demandas Marshallianas.

$$\frac{\frac{\partial V}{\partial w}}{\frac{\partial V}{\partial y}} = \frac{\lambda h}{\lambda} = h^*[T-l^*(P,w,y, \epsilon, A)]$$
