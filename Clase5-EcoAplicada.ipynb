##ENFOQUE PRECIO SOMBRA
Heckman (1974)

(Idea) el precio sombra permite:

    a) Modelar la decisión de entrar o no entrar a trabajar;
    b) ¿Cuántas horas?;

El salario de reserva se define como: (mujer casada)

$$w^* = g(h, w_m, \mathbf{P}, A, \mathbf{Z}),$$
donde,
$h \equiv$ horas trabajadas - tiempo que no cuenta para ocio;

$w_m \equiv$ salario de mercado del cónyuge;

$\mathbf{P} \equiv$ vector de precios;

$A \equiv$ riqueza: activos del hogar;

$\mathbf{Z} \equiv$ otras características: número de hijos, edad, etc;

Salario que define el mercado:

$$w = B(E, S),$$
donde $E$ representa a la experiencia y $S$ representa a la escolaridad.

Si $w^* > w |_{h=0}$, entonces decide no trabajar;

Si $w^*<w |_{h=0}$, entonces decide entrar a trabajar

El individuo entrará a la fuerza laboral y ofrecerá $h^*$, donde $w = w^*$. En el siguiente gráfico, $w_1$ es el precio de reserva.

##Estimación

Define forma funcional:

   [asking wage][4] $$\ln(w_i^*) = \beta_0 + \beta_1h_i + \beta_2w_m + \boldsymbol\beta_3\mathbf{P} + \beta_4 A_i + \boldsymbol\beta_5\mathbf{Z} + \boldsymbol\varepsilon$$

[offered wage][5]$$\ln(w_i) = b_0 + b_1S_i + b_2E_i + u_i$$

Asumimos que $\varepsilon_i$ & $u_i \sim \mathbf{MN}[\mathbf{0}, \Sigma]$

Sabemos por [3] lo siguiente:
   si  $h_i>0 \Longrightarrow \ln(w_i) = \ln(w_i^*)$;

   si  $h_i=0 \Longrightarrow \ln(w_i) < \ln(w_i^*)$;
   
   Por lo anterior, para los individuos que decidieron trabajar, sabemos que en $h=0$, $w_i > w_i^*$ ($w_i-w_i^* \equiv k > 0$).
   Así, 
   
   $$k = (b_0-\beta_0) + b_1S_i + b_2E_i - \beta_2 w_m - \boldsymbol\beta_3\mathbf{P} - \beta_4 A_i - \boldsymbol\beta_5\mathbf{Z} + \boldsymbol\varepsilon + u_i \Longrightarrow (b_0-\beta_0) + b_1S_i + b_2E_i - \beta_2 w_m - \boldsymbol\beta_3\mathbf{P} - \beta_4 A_i - \boldsymbol\beta_5\mathbf{Z}    > \boldsymbol\varepsilon - u_i$$

y trabajarán hasta que $\ln(w_i^*) = \ln(w_i)$. Para un salario $w_i$,

[asking wage] [7]$$h_i = \frac{1}{\beta_1}(b_0-\beta_0 + b_1S_i + b_2E_i - \beta_2 w_m - \boldsymbol\beta_3\mathbf{P} - \beta_4 A_i - \boldsymbol\beta_5\mathbf{Z}) + \frac{u_i - \epsilon_i}{\beta_1}$$

[offered wage][8]$$\ln(w_i)  = b_0 + b_1S_i + b_2E_i + u_i$$

El problema en [7] y [8] sólo se podría calcular para todos los que trabajan. Por lo anterior, el modelo no se puede estimar mediante MCO, ya que tenemos un problema de seleeción de muestra. Por lo anterior, planteamos una función de densidad condicional.

[9]$$f(h_i, w_i |w_i^* < w_i |_{h_i = 0}) = f(h_i, w_i |\varepsilon_i - u_i < k) = \frac{f(h_i, w_i)}{\mathbb{P}(\varepsilon_i - u_i < k)},$$
por Bayes.

Por lo anterior, Heckman propone estimar

[10] $$L = \prod_{h_i > 0} f(h_i, w_i | \varepsilon_i - u_i < k)·\mathbb{P}(\varepsilon_i - u_i < k)· \prod_{h_i =0}\mathbb{P}(k< \varepsilon_i - u_i),$$
donde el primer término es una función de distribución de $h_i$ y $w_i$; el segundo, la probabilidad de decidir trabajar; el tercero, la probabilidad de decidir no trabajar.

La estimación de [10] por MLE nos arrojará estimadores consistentes, insesgados y eficientes.

- ¿Cuál es la similitud con tobit? (Tobit solamente funciona para muestras censuradas (para los que tenían observaciones y no observaciones, tenían que ser las mismas 'x's').La similitud es que introduce la información de los que no trabajan, por lo que balancea el problema de selección de muestra. La diferencia del Heckit al tobit es que Heckman tuvo que usar un modelo estructural.
- Ventajas sobre tobit:
    * no asume las mismas x's y el mismo impacto para la decisión de trabajar y horas trabajadas;
    * Puede incorporar $w$ porque puede trabajar con muestra truncada no requiere las mismas x's.
