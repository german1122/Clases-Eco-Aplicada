Función de Utilidad directa:
[1]$$ \underset{\{x>0, l>0\}}{\mathrm{Máx}} \quad U = U[x, l; A, \epsilon],$$
s.a:
[4]$$px = wh + y, h>0$$
$$l = T- h$$

donde 
$x \equiv$ consumo  
$h \equiv$ horas trabajadas    
$T \equiv$ tiempo total disponible    
$l  = T-h\equiv$ ocio  
$A \equiv$ características observables del individuo   
$\epsilon \equiv$ características no observables del individuo   
$p \equiv$ precios   
$x \equiv$ bienes   
$w \equiv$ salario  
$y \equiv$ ingreso no laboral

[5]$$\mathcal{L} = U(x, l; A, \epsilon) + \lambda(wT + y -px-wl), x, h \in \mathbb{R_{++}}$$

Foc:
[6] $$\frac{\partial \mathcal{L}}{\partial x} = U_x -\lambda p = 0 \Longrightarrow U_x = \lambda p;$$

[7] $$\frac{\partial \mathcal{L}}{\partial l} = U_l -\lambda w = 0 \Longrightarrow U_w = \lambda w;$$

[8] $$\frac{\partial \mathcal{L}}{\partial \lambda} = wT + y -px -wl = 0$$

Así, $ \frac{U_l}{U_x} = \frac{w}{p} \equiv$ Tasa Marginal de Sustitución.

##Oferta Laboral
###Demanda Marshalliana

A partir de (6) y (7), aplicando (8):

[8]$$\frac{U_l}{U_x} = \frac{w}{p} \equiv TMS$$
$$\Longrightarrow l^* = l(p, w, y; A, \epsilon)$$
$$x^* = x(p, w, y; A, \epsilon),$$

donde $p,w, y;A$ son exógenos. Note que $\mathbb{E}[\epsilon | y] \neq 0$



##Propiedades de las demandas Marshallianas

Diferenciando totalmente [6] , [7] & [8], las condiciones de primer orden:

$$U_{xx}dx + U_{xl}dl - pd\lambda -\lambda dp = 0$$

$$U_{lx}dx + U_{ll}dl - wd\lambda -\lambda dw = 0$$

$$-pdx - wdl + 0d\lambda + (T-l)dw + dy - xdp = 0$$


[11]$$\begin{bmatrix}
U_{xx} & U_{xl} & -p \\[0.3em]
U_{lx} & U_{ll} & -w \\[0.3em]
p & w & 0
     \end{bmatrix} \begin{bmatrix}
\mathrm{d}x \\[0.3em]
\mathrm{d}l \\[0.3em]
\mathrm{d}\lambda
     \end{bmatrix}  = \begin{bmatrix}
\lambda \mathrm{d}p \\[0.3em]
\lambda \mathrm{d}w \\[0.3em]
(T-l)\mathrm{d}w + \mathrm{d}y -x\mathrm{d}p
     \end{bmatrix} \Longrightarrow 
     \begin{bmatrix}
\mathrm{d}x \\[0.3em]
\mathrm{d}l \\[0.3em]
\mathrm{d}\lambda
     \end{bmatrix} = \beta \begin{bmatrix}
\lambda \mathrm{d}p \\[0.3em]
\lambda \mathrm{d}w \\[0.3em]
(T-l)\mathrm{d}w + \mathrm{d}y -x\mathrm{d}p
     \end{bmatrix},$$ donde 
     
$$\beta = \begin{bmatrix}
b_{11} &  b_{12} & b_{13} \\[0.3em]
b_{21} & b_{22} & b_{23} \\[0.3em]
b_{31} &b_{31} & b_{33}
     \end{bmatrix}$$     




##Efecto del cambio en el ingreso no laboral

[14]$$\frac{\partial x}{\partial y} = b_{13}$$
[15]$$\frac{\partial l}{\partial y} = b_{23}$$


##Efecto de un cambio en el precio



$\Delta$ en precio

$$\frac{\partial x}{\partial p} = \lambda b_{11} - xb_{13}$$

Sustituyendo [14]

$$\frac{\partial x}{\partial p} = \lambda b_{11} - x\frac{\partial x}{\partial y}$$

El primer término de la ecuación anterior representa un efecto sustitución. Para entender el efecto sustitución, debemos preguntarnos cómo tienen que cambiar $p, w, y$ para mantener $U$ constante.

Sustituyendo [6] & [7],
$$\mathrm{d} U = U_{x}\mathrm{d}x + U_{l}\mathrm{d}l = \lambda p\mathrm{d}x + \lambda w \mathrm{d}l = \lambda(p\mathrm{d}x + w\mathrm{d}l)$$

$$A = p\mathrm{d}x + w\mathrm{d}l = \mathrm{d}y-x\mathrm{d}p+(T-l)\mathrm{d}w$$

$$0 = dU = \lambda[\mathrm{d}y - x\mathrm{d}p + (T-l)\mathrm{d}w]$$

Por el supuesto de no saciedad, $\lambda > 0$



Si sustituímos $0$ en [11], 

$$\begin{bmatrix}
\mathrm{d}x \\[0.3em]
\mathrm{d}l \\[0.3em]
\mathrm{d}\lambda
     \end{bmatrix} = \begin{bmatrix}
b_{11} &  b_{12} & b_{13} \\[0.3em]
b_{21} & b_{22} & b_{23} \\[0.3em]
b_{31} &b_{32} & b_{33}
     \end{bmatrix} \begin{bmatrix}
\lambda \mathrm{d}p \\[0.3em]
\lambda \mathrm{d}w \\[0.3em]
0
     \end{bmatrix}$$

[17]$$\frac{\partial x}{\partial p} = \frac{\partial x}{\partial p} \Bigg|_{dU = 0} = \lambda b_{11} $$



Sustituyendo [17] en [14], obtenemos la ecuación de Slutzky

[Ecuación de Slutsky]$$ \frac{\partial x}{\partial p} = \frac{\partial x}{ \partial p}\Bigg|_{dU = 0} - x \frac{\partial x}{\partial y},$$
donde el primer término es el efecto sustitución; el segundo, efecto ingreso. Sabemos que el efecto sustitución siempre será negativo.

##Oferta Laboral


$$\frac{\partial l}{\partial w} = \lambda b_{22} + b_{23}(T-l),$$

donde el segundo término es un efecto ingreso (no laboral) para un cambio en el salario. 

Sustituyendo [15] en lo anterior,

$$\frac{\partial l}{\partial w} = \frac{\partial l}{\partial w}\Bigg|_{dU = 0} + \frac{\partial l}{\partial y} (T-l)$$

Lo anterior es la ecuación de Slutsky para la demanda de ocio.

$$\frac{\partial l}{\partial w} = ES + EI$$

La regularidad empírica sugiere que el ocio es un bien normal, por lo que esperamos que el efecto ingreso sea positivo. El efecto sustitución siempre es negativo.
